```python
import os
import re
```

```python
files = os.listdir("../Business descriptions text/")
```

```python
def risk_index(contents):
    idx = -1
    x = re.findall("Item\s* 1a", contents, re.IGNORECASE)
    if len(x) > 0:
        idx = contents.index(x[0])
    return idx
```

```python
for f in files:
    contents = open(os.path.join("../Business descriptions text",f), "r").read()
    idx = risk_index(contents=contents)
    if idx > -1:
        print(idx)
        risk = contents[idx:]

        with open(os.path.join("../risk",f), "w") as outfile:
            outfile.write(risk)
```
